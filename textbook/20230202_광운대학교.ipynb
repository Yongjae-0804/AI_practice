```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([2, 4, 6, 8]) # 공부시간
x2 = np.array([1, 2, 1, 3]) # 문제집 푼 개수
y = np.array([20, 75, 55, 85])

lr = 0.005
a1 = 0
a2 = 0
b = 0

for i in range(3000):
    y_hat = a1 * x1 + a2 * x2 + b
    error = y_hat - y
    
    a1_diff = sum(2 * error * x1)
    a2_diff = sum(2 * error * x2)
    b_diff = sum(2 * error * 1)

    a1 = a1 - a1_diff * lr
    a2 = a2 - a2_diff * lr
    b = b - b_diff * lr

    print("{}회 학습 // 기울기1 : {} // 기울기2 : {} // 절편 : {}".format(i+1, a1, a2, b))
```

```python
plt.scatter(x1, y)
plt.plot(x1, a1*x1 + a2*x2 + b, 'red')
```

```python
# 사인 함수 예측하기
import numpy as np
import random

x = np.arange(-3, 3, 0.1)
y = np.sin(x)

# 오차가 가장 적은 a값과 b값을 찾기
lr = 0.00002
a = random.random() # 0~1 사이의 값
b = random.random()
c = random.random()
d = random.random()
e = random.random()

for i in range(20000):
    y_hat = a*x**4 + b*x**3 + c*x**2 + d*x + e
    error = y_hat - y

    a_diff = sum(2 * error * x**4)
    b_diff = sum(2 * error * x**3)
    c_diff = sum(2 * error * x**2)
    d_diff = sum(2 * error * x)
    e_diff = sum(2 * error * 1)

    a = a - a_diff * lr
    b = b - b_diff * lr
    c = c - c_diff * lr
    d = d - d_diff * lr
    e = e - e_diff * lr

    print("{}회 학습, a : {}, b : {}, c : {}, d : {}, e : {}".format(i+1, a, b, c, d, e))
```

```python
plt.plot(x, y)
plt.plot(x, a*x**4 + b*x**3 + c*x**2 + d*x + e, 'red')
```

```python
from tensorflow import keras # Google에서 만든 인공신경망 모듈

# 정형 데이터 : 머신러닝(랜덤포레스트, Xgboost)
# 비정형 데이터 : 인공신경망

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt

train_scaled = train_input.reshape(-1, 784) / 255
test_scaled = test_input.reshape(-1, 784) / 255
```

```python
from sklearn.ensemble import RandomForestClassifier

# 랜덤포레스트 : 87.66%
rf = RandomForestClassifier()
rf.fit(train_scaled, train_target)
rf.score(test_scaled, test_target)
```

```python
# 원핫인코딩
train_target = keras.utils.to_categorical(train_target)
test_target = keras.utils.to_categorical(test_target)
```

```python
# 은닉층의 활성화함수는 렐루하는게 좋음

model = keras.Sequential() # 도화지
model.add(keras.layers.Dense(256, activation = 'relu', input_shape = (784,)))
model.add(keras.layers.Dropout(0.4)) # 50%를 넘기면 안되요.
model.add(keras.layers.Dense(128, activation = 'relu'))
model.add(keras.layers.Dropout(0.4)) # 50%를 넘기면 안되요.
model.add(keras.layers.Dense(10, activation = 'softmax'))
model.summary()

# 10번 이상 연속으로 좋아지지 않으면 학습 강제종료 // 마지막 10번은 무효로 한다.
es = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)

model.compile(loss = "categorical_crossentropy", metrics = 'accuracy')

hist = model.fit(train_scaled, train_target, epochs = 10000,
                 validation_data = (test_scaled, test_target),
                 callbacks = [es])
```

```python
train_target
```

```python
plt.plot(hist.history['loss'][:-10])
plt.plot(hist.history['val_loss'][:-10])
plt.legend(['loss', 'val_loss'])
```

```python
# 최대 95% 92~93%

# 인공신경망 : 84.48%
# 딥러닝 : 89.41%
# 깊은딥러닝 87.8%

model.evaluate(test_scaled, test_target)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 인공신경망을 활용해서 iris 데이터를 학습하고, 평가해주세요.

import numpy as np

iris = load_iris()
data, target = iris['data'], iris['target']

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size = 0.2)

mean = np.mean(train_input, axis = 0)
std = np.std(train_input, axis = 0)

train_scaled = (train_input - mean) / std
test_scaled = (test_input - mean) / std

model = keras.Sequential()
model.add(keras.layers.Dense(8, activation = 'relu', input_shape = (4,)))
model.add(keras.layers.Dense(3, activation = 'softmax'))

# 15~30번 참기
es = keras.callbacks.EarlyStopping(patience=15, restore_best_weights=True)
model.compile(loss = "sparse_categorical_crossentropy", metrics = 'accuracy')

hist = model.fit(train_scaled, train_target, epochs = 10000,
                 validation_data = (test_scaled, test_target),
                 callbacks = [es])
```

```python
plt.plot(hist.history['loss'][:-15])
plt.plot(hist.history['val_loss'][:-15])
plt.legend(['loss', 'val_loss'])
```

```python
model.evaluate(test_scaled, test_target)
```
