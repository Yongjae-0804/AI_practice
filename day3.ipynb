```python
url = rq.get('https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C')
html = BeautifulSoup(url.text)

current = int(html.find('span',attrs={'class':'f_red'}).text[1:-1])

def annuity_lotto(html):
    team = html.find('td',class_='each').text # 문자열
    temp_num = html.find('div',class_ = "inner").text 
    lottonum = temp_num.split() 

    lottonum.insert(0,team)
    return lottonum
```

```python
total = []

for n in range(1,current+1):
    url = rq.get(f'https://search.daum.net/search?w=tot&DA=JIM&rtmaxcoll=JIM&&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C%20720%2B%20{n}%ED%9A%8C%EC%B0%A8')
    html = BeautifulSoup(url.text)
    num = annuity_lotto(html)
    print(f'{n}회 로또번호......{num}')
    time.sleep(0.7)
    total.append(num)

df = pd.DataFrame(total, columns = ['조','1','2','3','4','5','6'], index= range(1,current+1))
df.to_excel('연금복권.xlsx')
```

```python
add = 'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1'
url = rq.get(add)
html = BeautifulSoup(url.text)
data = html.find('table',class_= 'type_2')
kospi_page = int(html.find('td', class_= "pgRR").find('a')['href'][-2:])

#pd.read_html(str(data)) # list type 속에있는 것은 데이타 프레임
kospi_box = []

from tqdm import tqdm

for i in tqdm(range(1, kospi_page + 1)):
    url = rq.get(f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={i}')
    html = BeautifulSoup(url.text)
    data = html.find('table',class_= 'type_2')
    df = pd.read_html(str(data))[0] # 데이타 프레임
    df = df[df['종목명'].notnull()]
    del df['토론실']
    del df['N']

    kospi_box.append(df)
    time.sleep(0.7)

add = 'https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1'
url = rq.get(add)
html = BeautifulSoup(url.text)
data = html.find('table',class_= 'type_2')
kosdak_page = int(html.find('td', class_= "pgRR").find('a')['href'][-2:])

kosdak_box=[]
for i in tqdm(range(1, kosdak_page + 1)):
    url = rq.get(f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={i}')
    html = BeautifulSoup(url.text)
    data = html.find('table',class_= 'type_2')
    df = pd.read_html(str(data))[0] # 데이타 프레임
    df = df[df['종목명'].notnull()]
    del df['토론실']
    del df['N']

    kosdak_box.append(df)
    time.sleep(0.7)

kospi = pd.concat(kospi_box, ignore_index=True)
kospi['sosok'] = 'kospi'

kosdak = pd.concat(kosdak_box, ignore_index=True)
kosdak['sosok'] = 'kosdak'

stock = pd.concat([kospi,kosdak])
stock.index = stock.index+1 # 인덱스 다루기
stock.to_excel('stock.xlsx')
```

```python
df = pd.DataFrame(total, columns=['tiele','sub'])
df = df.drop_duplicates()
df
```

```python
import re

t = ''
for i in df['tiele']:
    t += i
    t += ' '
for i in df['sub']:
    t += i
    t += ' '
```

```python
words = re.findall('[a-zA-Z0-9가-힣%]+',t)
```

```python
dic = {}
for i in words:
    if i not in dic:
        dic[i] = 1
    else:
        dic[i] += 1

freq_df = pd.DataFrame({
    'word' : dic.keys(),
    'freq' : dic.values()
})

freq_df.sort_values(by = 'freq', ascending= False)
```
