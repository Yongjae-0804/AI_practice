```python
data = []

for i in range(2010,2021):
    for j in range(1,13):
        try:
            year = str(i)
            month = str(j).zfill(2)
            df = pd.read_excel(f'./data/tour/kto_{year+month}.xlsx', header=1, usecols='A:G', skipfooter=4)
            df['tour percent'] = df['관광']/df['계']*100
            df['total per'] = (df['관광']/df['관광'].sum())*100
            df['year'] = year
            df['month'] = month
            df['std_Y&M'] = f'{year}-{month}'
            data.append(df)
        except: pass
```

```python
df = pd.concat(data, ignore_index=True)
df
```
